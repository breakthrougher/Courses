```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# 设置图像目录
image_dir='runs/train/exp/'  # 替换为你的图片目录

# 获取目录下的所有图片文件
images = [img for img in os.listdir(image_dir) if img.startswith(('train'))]
# 设置显示的图像数量
num_images_to_display = min(9, len(images))  # 只显示最多9张图像

# 创建子图
fig, axes = plt.subplots(1, num_images_to_display, figsize=(15, 5))

# 显示每张图像
for ax, img in zip(axes, images[:num_images_to_display]):
    img_path = os.path.join(image_dir, img)
    img_data = mpimg.imread(img_path)
    ax.imshow(img_data)
    ax.set_title(img)
    ax.axis('off')  # 不显示坐标轴

plt.tight_layout()
plt.show()
```
