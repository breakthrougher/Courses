```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# ËÆæÁΩÆÂõæÂÉèÁõÆÂΩï
image_dir='runs/train/exp/'  # ÊõøÊç¢‰∏∫‰Ω†ÁöÑÂõæÁâáÁõÆÂΩï

# Ëé∑ÂèñÁõÆÂΩï‰∏ãÁöÑÊâÄÊúâÂõæÁâáÊñá‰ª∂
images = [img for img in os.listdir(image_dir) if img.startswith(('train'))]
# ËÆæÁΩÆÊòæÁ§∫ÁöÑÂõæÂÉèÊï∞Èáè
num_images_to_display = min(9, len(images))  # Âè™ÊòæÁ§∫ÊúÄÂ§ö9Âº†ÂõæÂÉè

# ÂàõÂª∫Â≠êÂõæ
fig, axes = plt.subplots(1, num_images_to_display, figsize=(15, 5))

# ÊòæÁ§∫ÊØèÂº†ÂõæÂÉè
for ax, img in zip(axes, images[:num_images_to_display]):
    img_path = os.path.join(image_dir, img)
    img_data = mpimg.imread(img_path)
    ax.imshow(img_data)
    ax.set_title(img)
    ax.axis('off')  # ‰∏çÊòæÁ§∫ÂùêÊ†áËΩ¥

plt.tight_layout()
plt.show()
```
